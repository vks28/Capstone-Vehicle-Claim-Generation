```python
import pandas as pd

# load your df which consists the images and the metadata (caption of the images)
df = pd.read_excel("")

df
```

```python
# identify where the lora weight file and the desired generated image folder is located at
lora_path = ""
gen_image_path = ""
```

```python
from diffusers import AutoPipelineForText2Image
from torch import manual_seed
import torch
import os
```

```python
# below codes take all the captions in the file and use generate images using those captions and lora weights
# model can be changed depending on the weight eg. sdxl1.5
pipeline = AutoPipelineForText2Image.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16).to("cuda")
pipeline.load_lora_weights(lora_path)

def dummy_checker(images, **kwargs):
    return images, [False] * len(images)

pipeline.safety_checker = dummy_checker

i = 1
for index, row in df.iterrows():
    prompt = row["Modified Captions"]
    image_name = row["Image"]

    generator = torch.Generator(device="cuda")
    # generator = torch.Generator(device='cpu'), not recommended
    generator.manual_seed(42)

    image = pipeline(prompt, generator=generator).images[0]

    save_path = os.path.join(gen_image_path, image_name)
    image.save(save_path)

    print(i)
    i = i+1
```
